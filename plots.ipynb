```python
from matplotlib import pyplot as plt
from PIL import Image
```

```python
%matplotlib inline
image = Image.open('tl1.jpg')
image2 = Image.open('tl2.jpg')
fig = plt.figure(figsize=(24, 13.5))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(image)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(image2)
```
